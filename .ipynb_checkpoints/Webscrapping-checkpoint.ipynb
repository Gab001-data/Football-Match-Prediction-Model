```python
from bs4 import BeautifulSoup
import timer 
import requests
import pandas as pd
```

```python
standingsUrl = 'https://fbref.com/en/comps/9/Premier-League-Stats'
data = requests.get(standingsUrl)
```

```python
soup = BeautifulSoup(data.text)
standings = soup.select('table.stats_table')[0]
links = standings.find_all('a')
links = [link.get('href') for link in links]
squad_links = [f'https://fbref.com{link}' for link in links if 'squads' in link]
squad_links
```

```python
#get arsenal match log
team_url = squad_links[0]
match_req = requests.get(team_url)
```

```python
# from io import StringIO
# match_data = StringIO(match_req.text)
match_df = pd.read_html(match_req.text, match="Scores & Fixtures")
#filter premier league games
match_df = match_df[0][match_df[0]['Comp']=='Premier League']
match_df
```

```python
#get shooting stats
shot_data = BeautifulSoup(match_req.text)
link_tags = shot_data.find_all('a')
links = [link.get('href') for link in link_tags]
shot_links = [f'https://fbref.com{l}' for l in links if l and 'all_comps/shooting' in l]
```

```python
#get shooting stats
shooting_url = shot_links[0]
shooting_req = requests.get(shooting_url)
```

```python
shooting_df = pd.read_html(shooting_req.text, match="Shooting")[0]
shooting_df.columns =shooting_df.columns.droplevel(0)
shooting_df
```

```python
# merge match log and shoting stats
match_stats = match_df.merge(shooting_df[['Date','Opponent','Gls','Sh','SoT','SoT%','G/Sh','G/SoT','Dist']], on=['Date','Opponent'], how='left')
match_stats.head()
```
