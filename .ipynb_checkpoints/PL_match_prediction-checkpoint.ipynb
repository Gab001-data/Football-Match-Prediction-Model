```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
#import data
match_df = pd.read_csv('PL_matches2023_2024season.csv')
match_df.drop('Unnamed: 0', axis=1, inplace=True)
match_df.head()
```

```python
#convert data types
match_df['Date'] = pd.to_datetime(match_df['Date'])
```

```python
#check for missing data
match_df.isna().sum()
```

```python
match_df[match_df.Result.isna()]
```

```python
#drop missing data
match_df2 = match_df.drop('Notes', axis=1).dropna(axis=0)
match_df2.isna().sum()
```

```python
#check record count
match_df2.shape
```

```python
#standardize team name
opponents = set(match_df['Opponent'])
teams = set(match_df['team'])
#get mismatching names
mismatch_names = []
for team in teams:
    if team not in opponents:
        mismatch_names.append(team)
mismatch_names
standard_names = {
    'BrightonandHoveAlbion':'Brighton',
    'WolverhamptonWanderers':'Wolves',
    'ManchesterUnited':'Manchester Utd',
    'NewcastleUnited':'Newcastle Utd',
    'AstonVilla':'Aston Villa',  
    'CrystalPalace':'Crystal Palace',    
    'WestHamUnited':'West Ham',  
    'LutonTown':'Luton Town',  
    'ManchesterCity':'Manchester City',      
    'NottinghamForest':"Nott'ham Forest",      
    'SheffieldUnited':'Sheffield Utd',  
    'TottenhamHotspur': 'Tottenham'                 
}
#standardize team names
match_df2['team'] = match_df2['team'].apply(lambda x:standard_names.get(x,x))
match_df2.head()
```

```python
#feature engineering
##generate encoding for categorical variables for modelling purpose

#get day of week
match_df2['Dayofweek'] = match_df2['Date'].dt.dayofweek

#get time of daty
match_df2['Timeofday'] = match_df2['Time'].str.replace(':\d+','',regex=True).astype('int64')

#generate venue code
encoder = LabelEncoder()
match_df2['Venue_code'] = encoder.fit_transform(match_df2['Venue'])

#generate encoding for teams
team_code = {name:code for code,name in enumerate(set(match_df2['team']))}

last_team_code = len(team_code)-1

#function to assign team code to related teams from previous seasons
def assign_code(team):
    """
        Asigns and return team code where not available in previous encoding
        due to unavailability of data (promotion/relegation)
        Args:
            team(str): team name
        Returns:
            int denoting team name code
    """
    if team not in team_code:
        global last_team_code
        last_team_code += 1
        team_code[team] = last_team_code
        return last_team_code
    else:
        return team_code[team]
match_df2['Opponent_code'] = match_df2['Opponent'].apply(assign_code)
match_df2['team_code'] = match_df2['team'].apply(lambda x: team_code[x])
```

```python
#check that teams relegated the previous season have been assign codes
match_df2[match_df2['Opponent_code']>19]
```

```python
#univariate analysis
##distribution of goals scored
g = sns.FacetGrid(match_df, col='team', col_wrap=5, sharex=False)
g.map(sns.histplot, 'GF', bins=6)
plt.show()
```

```python
#top match outcome by teams.
wins_df = match_df2[match_df['Result']=='W']
outcome_group = wins_df.groupby(by=['year','team','Result']).agg({'Result':'count','GF':'mean','GA':'mean','Sh':'mean','SoT':'mean','Dist':'mean'})
outcome_group.rename(columns={'Result':'Count'}, inplace=True)
outcome_group.reset_index()
outcome_group.sort_values(by=['year','Count','team'], ascending=False, inplace=True)
outcome_group.reset_index(inplace=True)
```

```python
#plot top teams with most wins per season
plt.figure(figsize=(10,8))
sns.barplot(outcome_group,x='Count', y='team', hue='year',orient='h')
plt.xlabel('Win Count')
plt.title('Count of Wins per Season')
plt.show()
```

```python
#bivariate relations 
plt.figure(figsize=(15,10))
sns.pairplot(match_df2[['GF','GA','Sh','SoT','Dist']])
plt.show()
```

```python
#multivariate relationship
results_df = match_df2.groupby(by=['year','team']).agg({'Result':'count','GF':'mean','GA':'mean','Sh':'mean','SoT':'mean','Dist':'mean'})
results_df.rename(columns={'Result':'Count'}, inplace=True)
results_df.reset_index(inplace=True)
results_df
```

```python
# Create scatterplot
df = results_df[results_df['year']==2023].reset_index(drop=True)
plt.figure(figsize=(10,8))
scatter_plot = sns.scatterplot(x='Sh',y='GF', hue='team', data=df, s=100, legend=False)

# Annotate each point with the category
for line in range(0, df.shape[0]):
     scatter_plot.text(df.Sh[line]+0.1, df.GF[line], df.team[line], horizontalalignment='left', size='medium', color='black', weight='semibold')

# Show plot
plt.title('Avg shots vs goal scored per teams')
plt.xlabel('Avg shots for')
plt.ylabel('Avg goals')
#plt.legend(title='Category')
plt.show()
```
