```python
#define function to scrape data for multiple seasons
def get_PL_data(data):
    seasons = list(range(2024,2022,-1))
    all_match_stats = []

    #get team urls
    soup = BeautifulSoup(data.text)
    standings = soup.select('table.stats_table')[0]
    links = standings.find_all('a')
    links = [link.get('href') for link in links]
    squad_links = [f'https://fbref.com{link}' for link in links if 'squads' in link]
    for team_url in squad_links:
        team_name = team_url.split('/')[-1].replace('-Stats','').replace('-','')
        for season in seasons:
            year = season
            #get team match log
            match_req = requests.get(team_url)
            match_df = pd.read_html(match_req.text, match="Scores & Fixtures")
            #filter premier league games
            match_df = match_df[0][match_df[0]['Comp']=='Premier League']
            #print(match_df)
            #get shooting link
            shot_data = BeautifulSoup(match_req.text)
            link_tags = shot_data.find_all('a')
            links = [link.get('href') for link in link_tags]
            shot_url = [f'https://fbref.com{l}' for l in links if l and 'all_comps/shooting' in l][0]
            #get shooting stats
            shooting_req = requests.get(shot_url)
            shooting_df = pd.read_html(shooting_req.text, match="Shooting")[0]
            shooting_df.columns =shooting_df.columns.droplevel(0)
            #print(shooting_df)
            # merge match log and shoting stats
            match_stats = match_df.merge(shooting_df[['Date','Opponent','Gls','Sh','SoT','SoT%','G/Sh','G/SoT','Dist']],\
                                         on=['Date','Opponent'], how='left')
            #print(match_stats)
            match_stats['team'] = team_name
            match_stats['year'] = season
            all_match_stats.append(match_stats)
            #get previous season match stats url for the team
            prev_data = BeautifulSoup(match_req.text)
            link_tags = shot_data.select('div.prevnext')
            prev_url =link_tags[0].find('a').get('href')
            team_url = f'https://fbref.com{prev_url}'

            time.sleep(1)
            
    return all_match_stats
        
    
```

```python
all_matches = get_PL_data(data)
```

```python
match_history_df = pd.concat(all_matches)
```
